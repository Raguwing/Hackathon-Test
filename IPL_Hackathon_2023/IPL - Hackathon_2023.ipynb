### 13.  What is the strike rate of Kohli in 2016


```python
final_df.head()
```

```python
total_runs = sum(final_df[(final_df.batsman == 'V Kohli') & (final_df.season == 2016)].total_runs)
```

```python
total_balls = sum(final_df[(final_df.batsman == 'V Kohli') & (final_df.season == 2016)].ball)
```

```python
strike_rate = (total_runs/total_balls)*100
```

```python
strike_rate
```
